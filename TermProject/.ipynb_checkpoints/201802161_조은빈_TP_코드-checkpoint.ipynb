```python
import os
import math
import numpy as np
import pandas as pd
import random
import shutil
from matplotlib import pyplot as plt
import scipy.io.wavfile as wav
import torch
import torchvision
from torch.optim.lr_scheduler import StepLR
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
def read_data():
    labels = []
    image_fns = []
    for f in os.listdir('./cats_dogs_sound'):   # 폴더에 있는 .wav 파일을 모두 읽는다
        if f.endswith('.wav'):
            base_fn = os.path.splitext(os.path.basename(f))[0] # .wav 파일 이름을 저장
            in_fn = os.path.join('./cats_dogs_sound', f)    # .wav 파일 경로 설정
            if not os.path.exists('./images'): os.makedirs('./images')
            to_fn = os.path.join('./images', base_fn+'.png')    # 스펙토그램 저장할 .png 파일 경로 설정
            draw_spectogram_image(in_fn, to_fn)
            
            image_fns.append(to_fn) # .png 파일 경로를 image_fns에 저장
            label = base_fn.split('_')[0]   # cat_1.wav 형식으로 저장된 파일 이름에서 cat인지 dog인지 label을 따온다
            labels.append(label)    # label을 labels에 저장

    with open('./labels.txt', 'w', encoding='utf-8') as f:
        for label in labels:
            print(label, file=f)    # labels text 파일을 만든다

    with open('./image_list.txt', 'w', encoding='utf-8') as f:
        for fn in image_fns:
            print(fn, file=f)   # image list 파일을 만든다
```

```python
def draw_spectogram_image(in_fn, to_fn):    # 스펙토그램 그리는 함수
    sample_rate, samples = wav.read(in_fn)  # .wav 파일의 Sample rate와 Data를 읽어온다
    noverlap=16
    cmap='viridis'
    dim_x, dim_y = 64, 64   # 이미지 크기
    fig = plt.figure()
    fig.set_size_inches((dim_x/fig.get_dpi(), dim_y/fig.get_dpi()))
    ax = plt.Axes(fig, [0.,0.,1.,1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    ax.specgram(samples, cmap=cmap,Fs=2, noverlap=noverlap)
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())
    fig.savefig(to_fn, bbox_inches="tight", pad_inches=0)   # 이미지 파일 저장
```

```python
def make_train_test():
    image_list = []
    labels = []
    with open('./labels.txt', 'r', encoding='utf-8') as f:
        for line in f:
            line = line.rstrip()
            labels.append(line)

    with open('./image_list.txt', 'r', encoding='utf-8') as f:
        for line in f:
            line = line.rstrip()
            image_list.append(line)

    # 셔플링
    data = [(image_fn, label) for image_fn, label in zip(image_list, labels)]

    random.seed(42)
    random.shuffle(data)

    N = len(data)
    train_idx = int(N*0.7)
    _train_data = data[:train_idx]
    _test_data = data[train_idx:]

    # cat, dog 폴더 안에 각각에 해당하는 이미지가 존재하도록
    # train 데이터 형태를 조정한다
    train_data = {}
    for image_fn, label in _train_data:
        if label not in train_data: train_data[label] = []
        train_data[label].append(image_fn)

    # cat, dog 폴더 안에 각각의 이미지가 존재하도록
    # test 데이터 형태를 조정한다
    test_data = {}
    for image_fn, label in _test_data:
        if label not in test_data: test_data[label] = []
        test_data[label].append(image_fn)

    if not os.path.exists('./run_images'): os.makedirs('./run_images')

    for label, image_fns in train_data.items(): # 훈련 데이터로 쓸 이미지를 복사해서 저장
        to_dir = os.path.join('./run_images/train', label)
        if not os.path.exists(to_dir): os.makedirs(to_dir)
        for image_fn in image_fns:
            shutil.copy(image_fn, to_dir)

    for label, image_fns in test_data.items():  # 테스트 데이터로 쓸 이미지를 복사해서 저장
        to_dir = os.path.join('./run_images/test', label)
        if not os.path.exists(to_dir): os.makedirs(to_dir)
        for image_fn in image_fns:
            shutil.copy(image_fn, to_dir)
```

```python
def load_dataset(data_path):    # 데이터들을 일정한 batch_size로 쪼개서 볼 수 있게 해준다
    _dataset = torchvision.datasets.ImageFolder(
        root = data_path,
        transform = torchvision.transforms.ToTensor()
        )
    _loader = torch.utils.data.DataLoader(
        _dataset,
        batch_size=128,
        num_workers=0,
        shuffle=True
        )
    return _loader
```

```python
def load_model(model_fn, map_location=None):
    if map_location:
        return torch.load(model_fn, map_location=map_location)
    else:
        if torch.cuda.is_available():
            return torch.load(model_fn)
        else:
            return torch.load(model_fn, map_location='cpu')
```

```python
class CNN_ASR(nn.Module):   # CNN 네트워크
    def __init__(self):
        super(CNN_ASR, self).__init__()
        self.conv1 = nn.Conv2d(3,32,3,1)    # 필터 생성
        self.conv2 = nn.Conv2d(32,64,3,1)
        self.dropout1 = nn.Dropout2d(0.25)  # overfitting 예방
        self.dropout2 = nn.Dropout2d(0.5)
        self.fc1 = nn.Linear(57600, 128)    # 차원 축소
        self.fc2 = nn.Linear(128, 10)
    def forward(self, x):
        x = self.conv1(x)
        x = F.relu(x)
        x = self.conv2(x)
        x = F.relu(x)
        x = F.max_pool2d(x, 2)
        x = self.dropout1(x)
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.dropout2(x)
        x = self.fc2(x)
        output = F.log_softmax(x, dim=1)
        return output
```

```python
def train(model, data_loader):
    optimizer = optim.Adadelta(model.parameters(), lr=0.1)
    use_cuda = torch.cuda.is_available()
    device = torch.device("cuda" if use_cuda else "cpu")
    model.to(device)
    model.train()

    for epoch in range(20):
        for batch_idx, (data, target) in enumerate(data_loader):
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            output = model(data)
            loss = F.nll_loss(output, target)
            loss.backward()
            optimizer.step()
            if batch_idx % 10 == 0:
                print('train Epoch: {} \tLoss: {:.6f}'.format(
                    epoch, loss.item()))
    torch.save(model, './model.out')
```

```python
def test(model, test_loader):
    use_cuda = torch.cuda.is_available()
    device = torch.device("cuda" if use_cuda else "cpu")
    model.to(device)
    model.eval()

    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item()
            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
```

```python
if __name__ == '__main__':
    read_data()
    make_train_test()

    train_data_path = './run_images/train'
    train_data_loader = load_dataset(train_data_path)

    test_data_path = './run_images/test'
    test_data_loader = load_dataset(test_data_path)

    torch.manual_seed(42)

    model = CNN_ASR()
    train(model, train_data_loader)

    model = load_model('./model.out')
    test(model, test_data_loader)
```
